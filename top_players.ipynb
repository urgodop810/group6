```python
import numpy as np
import pandas as pd
```

```python
player_centralities = pd.read_csv('player_centralities.csv')
player_centralities
```

```python
player_xT = pd.read_csv('player_xT.csv').rename(columns= {'Unnamed: 0':'player','0':'xT'})
player_xT
```

```python
player_ranking = player_xT
player_ranking['mean_centrality'] = np.nan
for i,row in player_ranking.iterrows():
    if row.player in player_centralities.player.values:
        player_ranking.loc[i,'mean_centrality'] = player_centralities.loc[player_centralities.player == row.player,'mean'].values[0]
    else:
        player_ranking.loc[i,'mean_centrality'] = 0
player_ranking
```

```python
#Normalising both xT and mean_centrality to be [0,1]
player_ranking.xT = (player_ranking.xT - player_ranking.xT.min()) / (player_ranking.xT.max() - player_ranking.xT.min())
player_ranking.mean_centrality = (player_ranking.mean_centrality - player_ranking.mean_centrality.min()) / (player_ranking.mean_centrality.max() - player_ranking.mean_centrality.min())
player_ranking
```

```python
#Average score
player_ranking['score'] = (player_ranking.xT + player_ranking.mean_centrality)/2
player_ranking = player_ranking.sort_values(by='score',ascending =False)
top_5 = player_ranking.head(5)
top_5
```

```python
top_5.to_excel('player_ranking.xlsx')
```
